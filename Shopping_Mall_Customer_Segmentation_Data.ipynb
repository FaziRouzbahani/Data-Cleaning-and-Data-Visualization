```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/FaziRouzbahani/Data-Cleaning-and-Data-Visualization/main/Shopping%20Mall%20Customer%20Segmentation%20Data%20.csv")
df.head()
```

# **Data Exploration**

```python
#Checking whether there is null values
df.isnull().sum()
```

```python
#checking if there is any duplicate data
df.duplicated().value_counts()
```

```python
df.info()
```

```python
#statistical summary of columns
df.describe(include="all")
```

```python
#How many people in the range of 20-30
df[(df["Age"]>20) & (df["Age"]<30)]
```

```python
#number of Female
df[df["Gender"]=="Female"]
```

```python
#all the female or any one with the annual income more than 80000
df[(df["Gender"]=="Female") | (df["Annual Income"]>80000)]
```

```python
#normalize annual income column and create a new column named normalized annual income
df["Normalized Annual Income"]=(df["Annual Income"]-df["Annual Income"].min())/(df["Annual Income"].max()-df["Annual Income"].min())
df.head()
```

```python
#dropping a column
df_original=df.drop(columns=["Normalized Annual Income"])
df_original.head()
```

# **Data Visualization**

```python
#Create a pie chart for Gender
import seaborn as sns
import matplotlib.pyplot as plt

#df["Gender"].plot(kind="pie",figsize=(10,6))
df["Gender"].value_counts().plot(kind="pie",figsize=(10,6))
```

```python
#Create a pie chart for Gender through GroupBy
df_gb=df_original.groupby(["Gender"]).count()
df_gb.plot(kind="pie",y="Customer ID")
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/FaziRouzbahani/Data-Cleaning-and-Data-Visualization/main/Unemployment_Canada_1976_present.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[df["Full-time employment"].isnull()]
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
df.describe()
```

```python
df.head()
```

```python
df1=df.sample(500,random_state=42)
```

```python
df1.info()
```

```python
df2=df1[["GEO","Unemployment rate"]].groupby('GEO').mean().reset_index()
df2.head()
df2.plot(kind="bar",x="GEO",y="Unemployment rate",title="Unemployment rate based on Province")
```

**Newfoundland and Labrador have the most unemployment rate based on 500 sample data**
